```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
plt.style.use("seaborn")
```

```python
r_1 = 1
r_2 = 0.5
alpha_1 = 2
alpha_2 = 3 
k_1 = 1
k_2 = 0.8



N1_values, N2_values = meshgrid(arange(0, 2, 0.1),arange(0, 2, 0.1))
x_values = N1_values[0]

N1_dot = r_1*N1_values*((k_1-N1_values-alpha_2*N2_values)/k_1) 
N2_dot = r_2*N2_values*((k_2-N2_values-alpha_1*N1_values)/k_2) 

N_1_star = k_2-alpha_1*x_values
N_2_star = (k_2-x_values)/alpha_1


fig,ax = plt.subplots(figsize = (16,9))

ax.plot(x_values,N_1_star,c = 'C1',lw = 2,label = r'dN2/dt=0')
ax.plot(x_values,N_2_star,c  = 'C2',label = r'dN1/dt=0')
ax.streamplot(N1_values,N2_values,N1_dot,N2_dot,density = 5,color = 'C0')
ax.set(ylim=(0,k_2),xlim = (0,k_1))
ax.set(title = "Lotka-Volterra Phase Flow Diagram")
plt.legend(loc=1,prop={'size':25})
plt.savefig('plots/lotka_voltera_phase_flow.png',dpi=300)
```

```python
x_values, y_values = meshgrid(arange(0, 2, 0.1),arange(0, 2, 0.1))
x_values = x_values[0]

alpha,beta,gamma,delta = (0.1,0.3,0.15,0.4)
#alpha,beta,gamma,delta = (0.1,0.1,0.1,0.1)

x_dot = alpha*y_values-beta*(x_values*y_values)
y_dot = gamma*x_values*y_values-delta*y_values

plt.streamplot(x_values,y_values,x_dot,y_dot,density=2)
```

```python
x_values, y_values = meshgrid(arange(0, 5, 0.1),arange(0, 5, 0.1))
x_values = x_values[0]

x_dot = x_values-(x_values*y_values)
y_dot = x_values*y_values-y_values

plt.streamplot(x_values,y_values,x_dot,y_dot,density=2)
```
